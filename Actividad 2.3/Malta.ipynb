```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#carga de archivo
df= pd.read_excel('Malta.xlsx')
df
```

```python
#verificacion informacion del DataFrame
df.info()
```

a) Los registros de 5 host name diferentes

```python
#Aplicamos filtro "O" y filtro por objeto con los 5 nombres difrentes que solicita en este caso Carol, Paul, Frank, Michael, Yvette
fil1= df[(df["host_name"] == "Paul") | (df["host_name"] == "Carol") | (df["host_name"] == "Michael") | (df["host_name"] == "Frank") | (df["host_name"] == "Yvette")]
fil1
```

b) Lps registros de los host que se hayan unido a Airbnb antes de año 2019

```python
#primero convertimos la columna host since a tipo datatime
df["host_since"]=pd.to_datetime(df["host_since"])
#Ahora aplicamos un filtro de comparacion para ver solo lo que se unieron antes del 2019
fil2= df[df["host_since"] < "01/01/2019"]
fil2
```

c)Los registros de los host que responden a más tardar en 1 dia

```python
#filtro o: muestra solo los que tienen un tiempo de respuesta de mas tardar un dia, en este caso un dia, algunas horas o una hora
fil3= df[(df["host_response_time"] == "within a day") | (df["host_response_time"] == "within a few hours") | (df["host_response_time"] == "within an hour") ]
fil3
```

d) Los registros que tengan al menos 5 accommodates

```python
#filtro por comparacion para obtener solo los que tengan al menos 5 accomomodates
fil4= df[(df["accommodates"] >= 5)]
fil4
```

e) Los regitros de los tipo de cuarto: Entire home y private room

```python
#filtro O para que nos muestre unicamente los los tipos de cuarto que mnos solicita Private room y Entire home
fil5=df[(df["property_type"] == "Entire home/aptme") | (df["property_type"] == "Private room")]
fil5
```

f) Los registros que a lo mucho cuenten con 4 camas

```python
#filtro por comparacion que nos muestra unicamente los que tiene 4 camas o menos
fil6= df[(df["beds"] <= 4)]
fil6
```

g) Los resgistros que tengan 1 baño o que sean superhost pero con la condicion de que todos sean instant_bookable

```python
#El filtro principal es un Y, combinado con un O que establece tres condiciones con tres variables; que tengan un baño o sean superhost, y que ademas sean instant_bookable
fil7= df[((df["bathrooms"] == 1) | (df["host_is_superhost"] == "t")) & (df["instant_bookable"] == "t")]
fil7
```

h) Los registros cuyo review_scores_cleanliness sea menor de 3 review_scores_rating sea mayor de 3

```python
#filtro y: muestras los registros cuyos review_scores_cleanliness sean menores a 3 y los review_scores_cleanliness sean mayores a
fil8= df[(df["review_scores_cleanliness"] < 3) & (df["review_scores_cleanliness"] > 3)]
fil8
```
