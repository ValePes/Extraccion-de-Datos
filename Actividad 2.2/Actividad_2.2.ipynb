```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#carga de archivo
df= pd.read_excel('Datos de Facturación.xlsx')
df
```

```python
#verificacion informacion del DataFrame
df.info()
```

1-CVE_CLPV (valores: 1000 a 2000)

```python
#Filtro y: se muestra la informacion consecutiva de los valores mayor a 1000 y menor a 2000
df["CVE_CLPV"]= pd.to_numeric(df["CVE_CLPV"], errors="coerce")
fil1= df[(df['CVE_CLPV'] >= 1000) & (df['CVE_CLPV'] <= 2000)]
fil1
```

2-CVE_VEND     (Todas las claves excepto “5” y “4”)

```python
#filtro Y con filto not: muestra todas las clves exepto la 5 y la 4
fil2= df[~(df["CVE_VEND"] == 4) & ~(df["CVE_VEND"] == 5)]
fil2
```

3-FECHA_ENT   (Fechas de “28/02/2022”)

```python
#filtro por objeto: extraeremos unicamente los valores de FECHA_ENT ques ean igual a 28/02/2022
# Convertir la columna de fecha al tipo datetime
df['FECHA_ENT'] = pd.to_datetime(df['FECHA_ENT'], errors='coerce', format='%d/%m/%Y %H:%M')
fil3= df[df['FECHA_ENT'].dt.date == pd.to_datetime("28/02/2022").date()]
fil3
```

4-CAN_TOT       (Cantidades menores a 5951.7)   o  STATUS (“E”)

```python
#filtro O: muestra los valores de la columna CAN_TOT menores que 5951.7 O con STATUS igual E
fil4=df[(df["CAN_TOT"] < 5951.7) | (df["STATUS"] == "E")]
fil4
```

5-Solo las columnas: CVE_DOC, FECHA_ENT, FECHA_VEN y CAN_TOT

```python
#filtro por columnas no consecutivas filtra unicamente las columnas CVE_DOC, FECHA_ENT, FECHA_VEN y CAN_TOT
fil5= df.iloc[ : , [0,6,7,9]]  
fil5
```

6-Solo las filas de 7001-7099

```python
#filtro por filas: se filtran las filas consecutivas de la 7001 a la 8000 (la primera empezando desde 0)
fil6= df.iloc[7001:8000,: ] 
fil6
```

7-Index= CVE_VEND (valores: 1, 2)  (columna: FECHAELAB)

```python
#Filtro de filas vs colas
#Paso 1: se indica que la columna CVE_VEND sera utilizada como indice
df1= pd.read_excel('Datos de Facturación.xlsx', index_col=3)
#paso 2: se filtran las filas con valores 1 y 2, en las columnas se filtra unicamente la columna FECHAELAB
fil7= df1.loc[[1, 2], ["FECHAELAB"]]
fil7
```
